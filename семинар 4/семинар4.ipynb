1.1 Изучите количество памяти с помощью matplotlib

```python
plt.figure(figsize=(8,6))
plt.hist(df['Memory_Amount'], bins=20)
plt.title('Количество памяти по ноутбукам')
plt.xlabel('Memory_Amount')
plt.ylabel('Laptop_Amount');
```

1.2 Изучите стоимость ноутбуков с помощью matplotlib

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Price_euros'], bins=20)
plt.title('стоимость ноутбуков')
plt.xlabel('Price_euros')
plt.ylabel('Laptop_Amount');
```

1.3 Изучите вес ноутбуков с помощью matplotlib

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Weight'], bins=50)
plt.title('вес ноутбуков')
plt.xlabel('Weight')
plt.ylabel('Laptop_Amount');
```

2.1 Изучите распределение типов носителя

```python
memory = df["Memory_Type"].value_counts()
memory
```

```python
plt.figure(figsize=(10,8))
plt.pie(memory.values, autopct="%1.1f%%", labels = memory.index)
plt.title('распределение типов носителя')
plt.legend(memory.index, loc='upper right');
```

2.2 Изучите распределение компаний производителей

```python
company = df["Company"].value_counts()
company
```

```python
plt.figure(figsize=(8,4))
plt.bar(company.index, company.values)
plt.title('распределение компаний производителей')
plt.xticks(rotation=40);
```

2.3 Изучите распределение операционной системы

```python
operSistem = df["OpSys"].value_counts()
operSistem 
```

```python
plt.figure(figsize=(8,4))
plt.bar(operSistem .index, operSistem .values)
plt.title('распределение операционной системы')
plt.xticks(rotation=40);
```

```python
plt.figure(figsize=(10,10))
explode =[0.2, 0.03, 0.04, 0.05, 0.07, 0.06, 0.05, 0.04, 0.5]
plt.pie(operSistem .values,autopct='%1.1f%%', labels=operSistem.index,explode=explode)
plt.title('распределение операционной системы')
plt.legend(operSistem.index, loc='best');
```

2.4 Изучите распределение компаний производителей CPU

```python
data = df["Cpu_Company"].value_counts()
data 
```

```python
plt.figure(figsize=(5,5))
explode =[0.05, 0.03, 0.07]
plt.pie(data .values,autopct='%1.1f%%', labels=data.index,explode=explode, colors=['0.3','g','#FF00BB'])
plt.title('распределение компаний производителей CPU')
plt.legend(data.index, loc='best');
```
