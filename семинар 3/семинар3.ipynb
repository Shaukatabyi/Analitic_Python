### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
df['Memory'].value_counts()
```

```python
df['Memory_Amount'] = df['Memory'].apply(lambda x: x.split()[0])
```

```python
def mem_value(x):
    if x.split()[0].endswith('GB'):
        res = x.split()[0].replace('GB','')
    elif x.split()[0].endswith('TB'):
        res = float(x.split()[0].replace('TB',''))*1024
    return int(res)
```

```python
def take_gb(text):
    if text.endswith('GB'):
        res =text.replace('GB', '')
    elif text.endswith('TB'):
        res = float(text.replace('TB', ''))*1024
    return int(res)
```

```python
df['Memory_Amount'] = df['Memory_Amount'].apply(lambda x: take_gb (x))
```

```python
df['Memory_Amount'].value_counts()
```

```python
df.head(2)
```
