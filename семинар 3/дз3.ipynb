#### Задача 1.2
Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома

Если реновации дома не было, то в новом признаке поставьте 0

```python
dom['delta_renovated'] = dom['yr_renovated']-dom['yr_built']
```

```python
dom['delta_renovated'] = dom['delta_renovated'].apply(lambda x: x if x>0 else 0)
```

```python
dom.head()
```

```python
dom[dom['yr_renovated'] > 0][['yr_built', 'yr_renovated', 'delta_renovated']].head()
```

### Условие 3: 
#### 3 задача
Составьте несколько сводных таблиц
#### 3.1 
Найдите среднюю стоимость домов в зависимости от количества спален

```python
dom.groupby('bedrooms').agg({'price':'mean'}).sort_values('price')
```

### 3.2 

Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
dom.groupby('condition').agg({'price':['min','max', 'mean']})
```

### 3.3 

Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
dom.pivot_table(index='view',
                columns= 'waterfront',
                values='id',                
                aggfunc='count',
                fill_value='-')
```

### 3.4 
Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=dom['floors'],
            columns= dom['bedrooms'])
```

### 3.5 
Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
pd.crosstab(index=dom['grade'],
            columns=dom['condition'],
            values=dom['price'],
            aggfunc='median',
            )
```

```python
dom.pivot_table(index='grade',
            columns='condition',
            values='price',
            aggfunc='median',
            fill_value='-')
```
